Import Libraries

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

Load Dataset

```python

data = pd.read_csv("spam.csv", encoding="latin-1")
```

Keep only required columns

```python
data = data[['v1', 'v2']].copy()
```

Rename columns

```python
data.columns = ['label', 'message']
```

Ensure label is STRING before cleaning

```python

data['label'] = data['label'].astype(str)
```

Clean labels

```python

data['label'] = data['label'].str.strip().str.lower()
```

Convert labels to numbers

```python
data['label'] = data['label'].map({'ham': 0, 'spam': 1})
```

Remove rows with missing values

```python
data = data.dropna()
```

Split Features & Target

```python
X = data['message']
y = data['label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

Text Vectorization (TF-IDF)

```python
vectorizer = TfidfVectorizer(stop_words='english')

X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

Train Model

```python
model = MultinomialNB()
model.fit(X_train_tfidf, y_train)
```

Evaluate Model

```python
y_pred = model.predict(X_test_tfidf)

print("Model Accuracy:", accuracy_score(y_test, y_pred))
```

Predict New Email

```python

def predict_email(email_text):
    email_tfidf = vectorizer.transform([email_text])
    prediction = model.predict(email_tfidf)

    return "ðŸš¨ Spam Email" if prediction[0] == 1 else "âœ… Not Spam Email"


# Example
sample_email = "Congratulations! You won a free prize. Click now!"
print("Prediction:", predict_email(sample_email))
```
